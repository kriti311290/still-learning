```python
import pandas as pd
import numpy as np
```

```python
import matplotlib as mp
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
x = np.linspace(0,5,11)
```

```python
x
```

```python
y = x**2
```

```python
y
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y,'r+')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('graph')
plt.show()
```

```python
plt.plot(x,y,'r+')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('graph')
plt.plot(y,x,'b+')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('graph')
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r--')
plt.subplot(1,2,2)
plt.plot(y, x,'g*-')
```

```python
plt.subplot(2,1,1)
plt.plot(x,y,'r--')
plt.subplot(2,1,2)
plt.plot(y, x,'g*-')
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y,'b')
axes.set_xlabel('set x label')
axes.set_ylabel('set y label')
axes.set_title('graph')
fig2 = plt.figure()
axes = fig2.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y,'r')
axes.set_xlabel('set x label')
axes.set_ylabel('set y label')
axes.set_title('graph')
fig.show()
fig2.show()
```

# data visualisation -> where to look, relation to previous samples


```python
fig= plt.figure()
axis1= fig.add_axes([0.1,0.1,0.8,0.8])
axis2= fig.add_axes([0.2,0.5,0.4,0.3])
axis1.plot(x,y,'b')
axis1.set_xlabel('set x label')
axis1.set_ylabel('set y label')
axis1.set_title('bada graph')
axis2.plot(y,x,'b')
```
